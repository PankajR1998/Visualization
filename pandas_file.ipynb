# Visualization using matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# boxplot represent the percentile topics.
data  = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

```python
# Pie-Chart
label = ['C++','Python','Ruby','Swift']
size = [215,160,410,200]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (.2,0,0,0)

plt.pie(size,labels=label,colors=colors,explode=explode, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.dtypes
```

```python
# here will find the correlation between the features.
df.corr()
```

```python
# here, it has found the correlation between the variable.
sns.heatmap(df.corr())
# Correlation is used for feature selection.
```

Joint Plot: it basically used to find the relationship between two variables. it shows scatterplot, hexbins plot, histogram plt.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')

# plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair-Plot helps in doing the multivariate analysis.

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data = df,hue='sex')
plt.savefig
```

## Dist-Plot: it is used to see the distribution of the given column.

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True)
```

# Plots for the Categorical features.

```python
# Count plot 
sns.countplot(x='sex',data = df)
plt.title('Count Plot for different Sex')
```

```python
sns.countplot(df['day'])
```

```python
sns.countplot(y='day',data=df,hue='sex')
```

```python
# Bar-Plot: it basically uses two variable to give us the plot.
sns.barplot(x='smoker',y='total_bill',data=df,hue='sex')
```

```python
# box-plot
sns.boxplot(data=df,x='smoker',y='total_bill')
plt.title('Categorical Analysis of smokers')
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(x='total_bill',y = 'day',hue='sex',data=df)
```

```python
# Violine-plot: it basically gives us the propertiy of box plot and the density view of data.
sns.violinplot(x='day',y='total_bill',data=df)
```
